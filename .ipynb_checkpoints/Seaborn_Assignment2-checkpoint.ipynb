# Q2: Create a histogram, where the x-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the day column's categories. Hint: use hue argument.

```python
# write your code here ^_^

sns.set_style('whitegrid')

sns.histplot(x='tip',hue='day',data=tips_dataset,element='step')
plt.show()
```

- Saturday tends to receive the highest amount of customers tips, possibly due to the restaurant usually being visited during weekends.
- Saturdays receive the highest amount of tips, with a score of 10

# Q4: Create a ecdf plot, where the x-axis represents the tip column
- Change the style to 'darkgrid'.
- Change the font scale to 1.25.
- Split the data out by the time column's categories. Hint: use hue argument.
- Change the lines color to 'summer'.
- Change the line width to 3.

```python
# write your code here ^_^
sns.set_style('darkgrid')
sns.set_context(font_scale=1.25)
sns.ecdfplot(x='tip',hue='time',palette='summer',linewidth=3,data=tips_dataset)
plt.show()
```

## Q4.1 Write your top three findings 
- Most high paid tips are at Dinner
- The graph demonstrates a gradual and consistent increase in tip amounts during both lunch and dinner periods.
- Based on the observation from the graph, it can be inferred that there is a notable concentration of tips in the range of 0.2 to 0.4, where a significant proportion of customers tend to pay exactly 2 units as tips.

# Q6: Create a violin plot, where the x-axis represents the time column and the y-axis represents the tip column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Split the violin so that each violin contains both sex column's categories. Hint: use split argument.
- Change the shape of the inner part of the violin to 'quartiles'.
- Change the violin color to 'Pastel1'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.
- Change the legend's location to The upper left.

```python
# write your code here ^_^
sns.violinplot(x='time',y='tip',hue='sex',split='sex',inner='quartiles',color='#FBB4AE',hue_order=['Female','Male'],linewidth=2,data=tips_dataset)
plt.legend(loc='upper left');

```

## Q6.1 Write your top three findings 
- the majority of the highest tips were observed during dinner service, with some reaching as high as 10. In contrast, the highest tips recorded during lunch service were relatively lower, with a maximum value of 8.
- Most females tend to pay 2 as a tip during lunch as it's demonstrated in the violin density
- As indicated by the right violin chart most men's tend to pay 3 at dinner as the density is more shifted towards 3

# Q8: Create a correlation heatmap 
- Adjust the figure's size to be (8,5).
- Change the color map to 'RdBu'.
- Add annotations to the heatmap.
- Set the minimum value to be -1.
- Set the maximum value to be 1.

```python
# write your code here ^_^
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

sns.heatmap(data=tips_dataset.corr(), cmap='RdBu', annot=True, vmin=-1, vmax=1)

plt.title('Correlation Heatmap')

plt.show()
```

## Q8.1 Write your top three findings 
- There is a moderately strong positive linear relationship (correlation coefficient of 0.68) between the total_bill and tip. As the total_bill increases, the tip amount tends to increase.
- The table size and tip exhibit a moderate positive linear relationship (correlation coefficient of 0.49). Larger table sizes are associated with slightly higher tip amounts.
- There is a moderate positive linear relationship (correlation coefficient of 0.6) between the total_bill and size. As the total_bill increases, the size of the dining party tends to increase.
